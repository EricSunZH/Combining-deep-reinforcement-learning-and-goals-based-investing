```python
import sys
import numpy as np
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
def read_data(data_file, DATA_PATH):
    df = pd.read_excel( os.path.join(DATA_PATH, data_file) )
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.set_index('Date') 
    df = df.loc['12/31/2001':,]
    
    return df
```

```python
data_file = 'BCOMTR.xlsx'
DATA_PATH = "./data"
data = read_data(data_file, DATA_PATH)
data = data.rename(columns={"PX_LAST": "BCOMTR"})
```

```python
indexes = np.array(['LBUSTRUU', 'LG30TRUU', 'LMBITR', 'NDDUE15X', 'NDDUJN', 'NDDUNA', 'NDDUPXJ', 
                    'NDDUUK', 'NDDUWXUS', 'NDUEEGF', 'RU10GRTR', 'RU10VATR', 'RU20GRTR', 'RU20INTR', 'RU20VATR',
                    'RUMCINTR', 'RUMRINTR', 'RUTPINTR', 'S5COND', 'S5CONS', 'S5ENRS', 'S5FINL', 'S5HLTH',
                    'S5INDU', 'S5INFT', 'S5MATR', 'S5RLST', 'S5TELS', 'S5UTIL', 'SPXT'])

for i in indexes:
    
    data_file = i+'.xlsx'
    DATA_PATH = "./data"
    df = read_data(data_file, DATA_PATH)
    
    data.insert(0, i, df['PX_LAST'], True)
```
